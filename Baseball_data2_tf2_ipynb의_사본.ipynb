```
from google.colab import drive
drive.mount('/content/drive')
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import random
```

```
xy = np.loadtxt('/content/drive/MyDrive/Colab_Notebooks/Baseball data/DATASET_TEST_V3.csv',delimiter=',',dtype = np.float32)

x_data = xy[:,0:-1]
y_data = xy[:,[-1]]
print(x_data.shape, y_data.shape)

learning_rate = 0.001
batch_size = 100
training_epochs = 500
nb_classes = 27
```

```
y_one_hot = tf.keras.utils.to_categorical(y_data, nb_classes)
print("one_hot:", y_one_hot)
```

```
tf.model = tf.keras.Sequential()

tf.model.add(tf.keras.layers.Dense(input_dim=42, units=20, kernel_initializer='glorot_normal', activation='relu'))
tf.model.add(tf.keras.layers.Dense(units=20, kernel_initializer='glorot_normal', activation='relu'))
tf.model.add(tf.keras.layers.Dense(units=nb_classes, kernel_initializer='glorot_normal', activation='softmax'))


tf.model.compile(loss='categorical_crossentropy', optimizer=tf.keras.optimizers.Adam(lr=learning_rate), metrics=['accuracy'])
tf.model.summary()
```

```
history = tf.model.fit(x_data, y_one_hot, epochs=training_epochs)
```

```
y_predicted = tf.model.predict(x_data)
for x in range(0, 100):
    random_index = random.randint(0, x_data.shape[0]-1)
    s=np.argsort(y_predicted[random_index])
    u=y_predicted[random_index][s]
    print("index: ", random_index,
          "\tactual y: ", np.argmax(y_one_hot[random_index]),
          "\tpredicted y: ", np.argmax(y_predicted[random_index])
          ,"\tpredicted best 3 score: ",np.flip(s)[:3],
           "\t" , np.flip(u)[:3] )

```

```
# evaluate test set
evaluation = tf.model.evaluate(x_data, y_one_hot)
print('loss: ', evaluation[0])
print('accuracy', evaluation[1])
```
